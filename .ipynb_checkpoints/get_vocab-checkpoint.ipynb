```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download NLTK data if not already available
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')

# Path to the file
file_path = r'extracted_passage.txt'

# 1. Read the file and store the content as a single string
with open(file_path, 'r') as file:
    file_content = file.read()

# 2. Convert the text to lowercase
file_content = file_content.lower()

# 3. Remove all punctuation and special characters using regex (keeping words and spaces only)
file_content = re.sub(r'[^\w\s]', '', file_content)

# 4. Remove numbers (e.g., years, dates)
file_content = re.sub(r'\b\d+\b', '', file_content)

# 5. Tokenize the text into words
words = word_tokenize(file_content)

# 6. Filter out stopwords (common words)
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words if word not in stop_words]

# 7. Use POS (Part of Speech) tagging to filter out proper nouns (NNP), numbers, etc.
tagged_words = nltk.pos_tag(filtered_words)

# 8. Keep only words that are not proper nouns, numbers, or other irrelevant tags
vocabulary_words = [word for word, pos in tagged_words if pos not in ['NNP', 'CD']]

# 9. Remove duplicates by converting the list to a set and back to a list
vocabulary_words = list(set(vocabulary_words))

# 10. Print the final vocabulary words
print(vocabulary_words)
```

```python
vocabulary_words_string = "['" + "', '".join(vocabulary_words) + "']"
print(vocabulary_words_string)

```

```python
file_path = 'vocab_sys_prompt.txt'
with open(file_path, 'r') as file:
    sys_prompt = file.read()  
    
print(sys_prompt)
print('-' * 50)

file_path = 'vocab_user_prompt.txt'
with open(file_path, 'r') as file:
    user_prompt = file.read()  
    
user_prompt = user_prompt + ' ' + vocabulary_words_string
print(user_prompt)
```

```python
import os
import base64
from together import Together
from PIL import Image
import io

# Initialize API client
client = Together(base_url="https://api.aimlapi.com/v1", api_key=API_key)

# Load image from your local drive
# THIS CODE IS NOT BEING USED BELOW
image_path = 'reading_test_3_passage_1/stitched_passage.png'  # Replace with your image path
with open(image_path, "rb") as image_file:
    # Convert the image to base64 string
    encoded_image = base64.b64encode(image_file.read()).decode('utf-8')

# print(encoded_image)
# print(type(encoded_image))
# Create the API request
response = client.chat.completions.create(
    model="meta-llama/Llama-3.2-90B-Vision-Instruct-Turbo",
    messages=[
            {
                "role": "system",
                "content": sys_prompt,
            },
        
            {
                "role": "user",
                "content": [
                    {
                        "type": "text",
                        "text":  user_prompt,
                    },
                    
#                     {
#                         "type": "image_url",
#                         "image_url": {
#                             # stitched passage.
#                             "url" :"https://drive.google.com/uc?id=1iQQM14fOhw6MOPxexPR03_u9GG7BDXTr"
# #                             "url": "https://drive.google.com/uc?id=1XM6uNKcw-kvX0i05D2N-eoHmIonXuTPx",
# #                             "url": "https://drive.google.com/uc?id=1GXOZ7spaWSkcvqMwXU6DGwr2IvBiNbjv",
#                         },
#                     },
                    
                    # IT IS NOT WORKING LIKE THIS, NOT TAKING MULTIPLE IMAGES AT ONCE. (MAYBE I AM MISSING SOMETHING)
#                     {
#                         "type": "image_url",
#                         "image_url": {
#                             # second page of passage.
#                             "url": "https://drive.google.com/uc?id=1GXOZ7spaWSkcvqMwXU6DGwr2IvBiNbjv",
#                         },
#                     },
                                
                ],
            }
        ],
    # IELTS PASSAGES ARE MOSTLY AROUND 2000-3000 WORDS
    max_tokens=3500,
)
```

```python
# Print the model's response
print("Assistant: ", response.choices[0].message.content)

response_text = response.choices[0].message.content
# Write the response to a .txt file
file_path = 'extracted_vocab.txt'  # Save it as a .txt file
with open(file_path, 'w') as file:
    file.write(response_text)  # Write the JSON response as plain text

print(f"Assistant response saved to {file_path}")
```
